```python
import numpy as np
import matplotlib.pyplot as plt

# Simple nonlinear function: f(x) = x[0]^2 + x[1]^4
def f(x):
    return x[0]**2 + x[1]**4

def gradient_f(x):
    return np.array([2*x[0], 4*x[1]**3])

x = np.array([2.0, 2.0])  # Starting point
gamma = .01  # Step size
iterations = 2000
path = [x.copy()]

for _ in range(iterations):
    x = x - gamma * gradient_f(x)  # Gradient descent step
    path.append(x.copy())

path = np.array(path)
plt.plot(path[:, 0], path[:, 1], '-o')
plt.xlabel('x1'); plt.ylabel('x2')
plt.title('Gradient Descent Path')
plt.show()
```

```python
import numpy as np
from numpy import arange
from matplotlib import pyplot
import matplotlib.pyplot as plt

# Simple nonlinear function: f(x) = x[0]^2 + x[1]^4
def f(x,y):
    return x**2 + y**4

r_min, r_max = -1.0, 1.0
inputs=arange(r_min, r_max, .1)
result=f(inputs,inputs)

pyplot.scatter(inputs, result)
plt.title('x**2 + y**4 graph')
plt.show()
```
